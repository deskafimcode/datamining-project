## Bagging

```python
import pandas as pd
```

```python
distilbert = pd.read_csv("data_komen_sampling20_pluslabel_distilbert.csv")
ditilbert = distilbert.rename(columns={"confidence_score" : "score"})
distilbert["sentiment_label"] = distilbert["sentiment_label"].map({"POSITIVE" : 1, "NEGATIVE" : 0})
distilbert.head()
```

```python
roberta = pd.read_csv("data_komen_sampling20_pluslabel_siebertroberta.csv")
roberta = roberta.rename(columns={"confidence_score" : "score"})
roberta["sentiment_label"] = roberta["sentiment_label"].map({"POSITIVE" : 1, "NEGATIVE" : 0})
roberta.head()
```

```python
bert = pd.read_csv("bert_vian.csv")
bert = bert.rename(columns={'bert_label': 'sentiment_label', "bert_score" : "score"})
bert.head()
```

```python
deberta_bagus = pd.read_csv("deberta_bagus.csv")
deberta_bagus = deberta_bagus.rename(columns={'deberta_label': 'sentiment_label', "deberta_score" : "score"})
deberta_bagus.head()
```

```python
roberta_bagus = pd.read_csv("roberta_bagus.csv")
roberta_bagus = roberta_bagus.rename(columns={'roberta_label': 'sentiment_label', "roberta_score" : "score"})
roberta_bagus.head()
```

```python
d_distilbert = distilbert.to_dict(orient='list')
d_roberta = roberta.to_dict(orient='list')
d_bert = bert.to_dict(orient='list')
d_deberta_bagus = deberta_bagus.to_dict(orient='list')
d_roberta_bagus = roberta_bagus.to_dict(orient='list')
```

```python
data_bert = [d_distilbert,d_roberta,d_bert,d_deberta_bagus,d_roberta_bagus]
label_bagging = []

for i in range(len(d_bert["review_detail"])) :
    positif = 0
    negatif = 0
    for j in data_bert :
        if j["sentiment_label"][i] == 1 :
            positif += 1
        else :
            negatif += 1
    if positif > negatif :
        label_bagging.append(1)
    else :
        label_bagging.append(0)
        
print(label_bagging)
print(len(label_bagging))
```

```python
data = pd.read_csv("data_sampling20.csv")
```

```python
data["label_komen"] = label_bagging
```
