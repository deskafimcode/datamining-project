## Klasifikasi Flop, Below Average, Average, Hits

```python
import pandas as pd

data = pd.read_csv("Hasil_Klasifikasi_Review.csv")
data.head()
```

```python
dict_rating_label = {}

for i in range(data.shape[0]) :
    if data["movie"][i] not in dict_rating_label :
        dict_rating_label[data["movie"][i]] = [data["rating"][i],data["label_komen"][i],1]
    else :
        dict_rating_label[data["movie"][i]][0] += data["rating"][i]
        dict_rating_label[data["movie"][i]][1] += data["label_komen"][i]
        dict_rating_label[data["movie"][i]][2] += 1
        
for i in dict_rating_label :
    dict_rating_label[i][0] /= dict_rating_label[i][2]
    dict_rating_label[i][1] /= dict_rating_label[i][2]

dict_hasil = {"movie" : [], "avg_rating" : [], "avg_sentiment" : [], "klasifikasi" : []}

for i in dict_rating_label :
    dict_hasil["movie"].append(i)
    dict_hasil["avg_rating"].append(dict_rating_label[i][0])
    dict_hasil["avg_sentiment"].append(dict_rating_label[i][1])
    if dict_rating_label[i][1] < 0.5 :
        if dict_rating_label[i][0] <= 5 :
            klasifikasi = "Flop"
        else :
            klasifikasi = "Average"
    else :
        if dict_rating_label[i][0] <= 5 :
            klasifikasi = "Below Average"
        else :
            klasifikasi = "Hits"
    dict_hasil["klasifikasi"].append(klasifikasi)
    
df_hasil = pd.DataFrame(dict_hasil)
df_hasil.head()
```

```python
df_hasil.to_csv("Klasifikasi 4 Kategori per Film.csv", index=False)
```
