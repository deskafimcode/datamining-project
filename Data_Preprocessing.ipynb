## Drop Null Value Rating

```python
data.dropna(subset=['rating'], inplace=True)
data = data.reset_index(drop=True)


ukuran_data = data.shape[0]
ukuran_data
```

## Reduksi Dimensi

```python
kolom_useless = ["reviewer","review_summary","review_date","spoiler_tag","helpful"]

for i in kolom_useless :
    del data[i]
    
data.head()
```

## Sampling 20%

```python
d_data = data.to_dict(orient="list")
d_data
```

```python
jumlah_film_min_pos = {}

for i in range(len(d_data["review_id"])) :
    if d_data["rating"][i] <= 5 :
        rate = "1-5"
    else :
        rate = "6-10"
    format_film = f"{rate} {d_data['movie'][i]}"
    
    if format_film not in jumlah_film_min_pos :
        jumlah_film_min_pos[format_film] = 1
    else :
        jumlah_film_min_pos[format_film] += 1
```

```python
data_sampling20 = {"review_id" : [], "movie" : [], "rating" : [] , "review_detail" : []}
counter_sampling = {}

for i in range(len(d_data["review_id"])) :
    if d_data["rating"][i] <= 5 :
        rate = "1-5"
    else :
        rate = "6-10"
    format_film = f"{rate} {d_data['movie'][i]}"
    
    if format_film not in counter_sampling :
        counter_sampling[format_film] = 1
        for j in d_data :
            data_sampling20[j].append(d_data[j][i])
    elif counter_sampling[format_film] < jumlah_film_min_pos[format_film]//5 :
        counter_sampling[format_film] += 1
        for j in d_data :
            data_sampling20[j].append(d_data[j][i])
    else :
        continue
    
df_data_sampling = pd.DataFrame(data_sampling20)

df_data_sampling.head()
```

```python
ukuran_data_sampling = df_data_sampling.shape[0]
ukuran_data_sampling
```

```python
ukuran_data_sampling/ukuran_data
```

```python
jumlah = {"0-5" : 0 , "5-10" : 0}

for i in d_data["rating"] :
    if i is None :
        continue
    if i < 5 :
        jumlah["0-5"] += 1
    elif i > 6 :
        jumlah["5-10"] += 1
        
print(jumlah)
neg1 = jumlah["0-5"]/(jumlah["0-5"]+jumlah["5-10"])
print(neg1)
```

```python
jumlah = {"0-5" : 0 , "5-10" : 0}

for i in data_sampling20["rating"] :
    if i is None :
        continue
    if i < 5 :
        jumlah["0-5"] += 1
    elif i > 6 :
        jumlah["5-10"] += 1
        
print(jumlah)
neg2 = jumlah["0-5"]/(jumlah["0-5"]+jumlah["5-10"])
print(neg2)
```

```python
persen_per_film = {}
for i in counter_sampling :
    persen_per_film[i] = counter_sampling[i]/jumlah_film_min_pos[i]
    
avg_persen_film = sum(list(persen_per_film.values()))/len(list(persen_per_film.values()))
avg_persen_film
```

## Memulai Labeling Negatif Positif BERT

```python
import torch
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
from datetime import datetime
import os

device = 0 if torch.cuda.is_available() else -1
print(f"Proses dijalankan menggunakan: {'GPU' if device == 0 else 'CPU'}")

pretrained_model = "distilbert-base-uncased-finetuned-sst-2-english"

nlp_detector = pipeline("sentiment-analysis", model=pretrained_model, device=device, batch_size=8)

list_komentar_anda = list(df_data_sampling["review_detail"])
output_csv = "data_komen_sampling20_pluslabel_distilbert.csv"

chunk_size = 1000
total_data = len(list_komentar_anda)

print("Memulai pelabelan dengan metode manual slicing...")

waktu1 = datetime.now()
print(f"NLPD Start : {waktu1}")

for i in tqdm(range(0, total_data, chunk_size)):
    potongan_list = list_komentar_anda[i : i + chunk_size]
    
    hasil_potongan = nlp_detector(potongan_list, truncation=True, max_length=512)
    
    data_buffer = []
    for teks, res in zip(potongan_list, hasil_potongan):
        data_buffer.append({
            'review_detail': teks,
            'sentiment_label': res['label'],
            'confidence_score': res['score']
        })
    
    df_chunk = pd.DataFrame(data_buffer)
    if not os.path.isfile(output_csv):
        df_chunk.to_csv(output_csv, index=False)
    else:
        df_chunk.to_csv(output_csv, mode='a', header=False, index=False)
        
    torch.cuda.empty_cache()

waktu2 = datetime.now()
print(f"\nNLPD Finish : {waktu2}")
print(f"Waktu Proses : {waktu2 - waktu1}")
print(f"Semua data berhasil dilabeli dan disimpan di: {output_csv}")
```
